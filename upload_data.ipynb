```python
# converting the .xslx file to .csv file for ease as the pd.read_Csv has more functionality than pd.read_excel 
# Note: While converting an Excel file to CSV, be aware that certain Excel features, such as multiple sheets, formulas, macros, and formatting, may not be preserved in the resulting CSV file, as CSV is a plain text format that does not support such features. Only the data from the Excel file will be included in the CSV file.

# Specify the path to the Excel file
excel_file = 'Financial_Sample.xlsx'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file)

# Specify the path for the CSV file
csv_file = 'Financial_Sample.csv'

# Convert the DataFrame to a CSV file
# The index=False argument ensures that the row index is not included in the CSV file.
df.to_csv(csv_file, index=False)
```

```python
# Create a Pandas dataframe with the first 100 rows of the csv
# See Pandas documentation or YouTube videos for info about dataframes
df = pd.read_csv('Financial_Sample.csv', nrows=100)
```

```python
# Show dataframe
df
```

```python
# Print schema
print(pd.io.sql.get_schema(df, name='Lloyds'))
```

```python
# Change schema so that the DATE columns is modelled with a datetime data type rather than text data type
df.Date = pd.to_datetime(df.Date)
```

```python
# Import create_engine from SQLalchemy
from sqlalchemy import create_engine
```

```python
# Create a SQLalchemy engine using the local Docker container server we have running at port 5431
engine = create_engine('postgresql://root:root@localhost:5431/Lloyds')
```

```python
# Connect to engine
engine.connect()
```

```python
# Create a dataframe iterator that will iterate over the csv in chunks of 100,000
# See Corey Schafer video above for info about iterators
df_iter = pd.read_csv('Financial_Sample.csv', iterator=True, chunksize=100_000)
```

```python
# Establish a while loop
while True:
    
    # This is an iterable - it's moving to the next iteration
    df = next(df_iter)
    
    # Change schema to reflect datetime (vs. text)
    df.Date = pd.to_datetime(df.Date)
    
    #Use the 'to_sql' method - connect via the engine. If a record already exists, append it
    df.to_sql(name='Lloyds', con=engine, if_exists='append')
    

    # Note - The code will end with an exception
```
